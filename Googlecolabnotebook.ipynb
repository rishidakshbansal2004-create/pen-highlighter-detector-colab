```python
import time
model = YOLO("/content/runs/detect/Pen_Highlighter_detector/weights/best.pt")

# Warmup
model("/content/IMG_20221209_124333_jpg.rf.RPWGJhTagBQEpNvYsRDv.jpg")

# Timing
times = []
for i in range(20):
    start = time.time()
    model("/content/IMG_20221209_124333_jpg.rf.RPWGJhTagBQEpNvYsRDv.jpg")
    end = time.time()
    times.append(end - start)

avg_time = sum(times) / len(times)
fps = 1 / avg_time

print(f"Average inference time: {avg_time*1000:.1f}ms")
print(f"FPS: {fps:.1f}")
```
