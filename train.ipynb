```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

results = model.train(
    data="/content/Highlighter-1/data.yaml",
    epochs=100,
    imgsz=640,
    batch=16,
    name="pen_highlighter_detector",
    device=0  
)
```
